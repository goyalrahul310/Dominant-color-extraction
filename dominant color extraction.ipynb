```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("l.jpg")
plt.imshow(img)
img.shape
```

```python
# convert bgr format to rgb
im = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(im)
plt.axis("off")
plt.show()
```

```python
print(im.shape)
#convert to linear form
im2 = im.reshape((-1,3)) 
```

```python
# use kmean
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=5)
```

```python
kmean.fit(im2)
```

```python
k= kmean.cluster_centers_
```

```python
k = np.array(k,dtype = 'uint8')
```

```python
k
```

```python
i = 1
color = []
for j in k:
    color.append(j)
    plt.axis('off')
    plt.subplot(1,5,i)
    i+=1
    a = np.zeros((150,100,3),dtype = 'uint8')
    a[:,:,:] = j
    plt.axis('off')
    plt.imshow(a)
```

```python
l = kmean.labels_
```

```python
new_img = np.zeros(((720* 800, 3)),dtype = 'uint8')
```

```python
for i in range(new_img.shape[0]):
    new_img[i] = color[l[i]]
```

```python
imk = new_img.reshape(720,800,3)
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.axis("off")
plt.imshow(im)

plt.subplot(1,2,2)
plt.axis("off")
plt.imshow(imk)

plt.show()
```
